## Dataset
The dataset for this project is one that contains information of about 10,000 movies from The Movie Database (TMDb) including ratings and revenues. The dataset can be downloaded [here](chrome-extension://efaidnbmnnnibpcajpcglclefindmkaj/https://s3.amazonaws.com/video.udacity-data.com/topher/2018/July/5b57919a_data-set-options/data-set-options.pdf) or from [Kaggle](https://www.kaggle.com/datasets/tmdb/tmdb-movie-metadata).

The focus of this project is to use the available data to establish the correlation between the revenues generated by movies against the effect of budgets, genres, release dates, directors and the production company of the various movies.

The result of the analysis after the data wrangling would be used to answer the following questions:

1. Is a movies likely to do well in the markets if it has a huge budget?
2. Do ratings affect the revenues of movies?
3. How have the revenues from movies changed over the years?



## Summary of Findings

From the investigations performed, it was observed that the most common genres amongst the top 20 highest revenue generating movies were Adventure, Science fiction, Action, Fantasy, and Thriller

Additionally, it was observed that there were three directors with more than one movie (Steven Spielberg with 3, James Cameron with 2 and Joss Whedon also with 2) in the top 20 and this may be an indication that directing may have an influence in how well a movie is received in the market.

Furthermore, the investigation revealed that there are certain production companies that have multiple movies in the top 20 but there also seem to be a lot of collaboration between them.

The scatter plot of budget against revenue for these top 20 movies does not show any correlation that is sufficient to draw any conclusions about this relationship.



## Conclusion

From the analysis performed, it can be concluded that there appears to be a strong correlation between the budget and ratings of movies to the revenues they generate in the market. Additionally, the investigation was able to establish that the revenues from movies have steadily increased over the years which is a direct indication that the movie industry is growing with even more opportunities for future growth.
